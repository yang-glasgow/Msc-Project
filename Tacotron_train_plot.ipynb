```python
import torch
import os

processed_loss = []
pretrained_loss = []

for epoch in range(10,561,10):
    checkpoint = torch.load(os.path.join('./processed_check_points/',f'checkpoint_epoch_{epoch}.pt'))
    processed_loss.append(checkpoint['loss'])

    checkpoint = torch.load(os.path.join('./pretrained_check_points/',f'checkpoint_epoch_{epoch}.pt'))
    pretrained_loss.append(checkpoint['loss'])
```

```python
import matplotlib.pyplot as plt

epochs = range(10, 561, 10)

plt.figure(figsize=(10, 6))

plt.plot(epochs, processed_loss, label='Training on initialised model', color='blue', marker='.')
plt.plot(epochs, pretrained_loss, label='Training on pretrained model', color='red', marker='.')

plt.title('Loss of two models every 10 epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.legend()
plt.savefig('tacotron2_train.png')
plt.show()
```
